```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
names = [
    'age',
    'work-class',
    'final_weight',
    'education',
    'education-num',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'cap-gain',
    'cap-loss',
    'hours-per-week',
    'native-country',
    'salary'
]
```

```python
df = pd.read_csv('data.txt', names=names)
df = df.loc[df['age'] != 'totally_legit_string']
df = df.replace(' ', '', regex=True)
```

```python
df.head(15)
```

Как видно, у нас есть "?". Давайте заменим их на NaN и посчитаем сколько у нас пустых значений.

```python
df = df.replace('?', np.nan)
```

```python
df.head(15)
```

```python
df.isnull().sum()
```

Давайте заменим их часто встречающимися значениями.

```python
for i in names:
    df[i].fillna(df[i].mode()[0], inplace=True)
df.isnull().sum()
```

```python
df.head(15)
```

Проверим есть ли у нас записи, где возраст меньше 17, если есть, удалим такие записи.

```python
df.loc[df['age'] < '17']
```

```python
df = df.loc[df['age'] >= '17']
```

Посмотрим на типы данных наших столбцов:

```python
df.dtypes
```

Переведем age, education-num и hours-per-week в float64.

```python
df['age'] = df['age'].astype(np.float64)
df['education-num'] = df['education-num'].astype(np.float64)
df['hours-per-week'] = df['hours-per-week'].astype(np.float64)
```

## Перекодировка категориальных значений

```python
from sklearn.preprocessing import LabelEncoder

category_cols = ['work-class', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex']
for i in category_cols:
    df[i] = LabelEncoder().fit_transform(df[i])
```

```python
df.head()
```

## Разбиваем датасет

Разобьём сперва на то что будем искать и из чего будем искать.

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

Тренировочные и тестовые сеты.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## Построение моделей

Логистическая регрессия.

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression(random_state=0)
lg.fit(X_train, y_train)
```

```python
y_lg = lg.predict(X_test)
print(np.concatenate((y_lg.reshape(len(y_lg), 1), y_test.reshape(len(y_test), 1)), 1))
```

Построим матрицу неточностей и посмотрим на точность логистической регрессии.

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_lg)
print(cm)
accuracy_score(y_test, y_lg)
```

Давайте попробуем увеличить точность предсказаний.

Random forest Classifier.

```python
from sklearn.ensemble import RandomForestClassifier

rgc = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)
rgc.fit(X_train, y_train)
```

```python
y_rgc = rgc.predict(X_test)
print(np.concatenate((y_rgc.reshape(len(y_rgc), 1), y_test.reshape(len(y_test), 1)), 1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix(y_test, y_rgc)
print(cm)
accuracy_score(y_test, y_rgc)
```

Нам удалось повысить точность.
